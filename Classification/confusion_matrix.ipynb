## ASSESSING A DIABETES PREDICTION CLASSIFIER USING  CONFUSION MATRIX

##### Confusion Matrix is a metric used for classification problems when accuracy is not sufficient to reflect the model's true performance.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

diabetes_df = pd.read_csv('diabetes_clean.csv')

diabetes_df
```

```python
X = diabetes_df.drop('diabetes', axis = 1).values
y = diabetes_df.diabetes

from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify = y, random_state = 20)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, classification_report
```

```python
knn= KNeighborsClassifier(n_neighbors=6)

knn.fit(X_train, y_train)

y_pred = knn.predict(X_test)

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

##### This shows a 69% precision rate for diabetes, that is, of all the predicted positive values for diabetes, 69% of the were actually correct, while 31% were false positives.

##### The recall rate is 54%, which means that only 54% of actual positive values for diabetes were predicted correctly.
